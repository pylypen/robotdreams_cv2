### Step 0

Run the necessary imports.

```python
import cv2
import dlib
# This lib needed only for google colab for displaying correct img
from google.colab.patches import cv2_imshow
```

### Step 1

Load an image (any image that contains faces).

```python
# Download image and convert it from BGR to RGB
img = cv2.imread('/content/face.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Step 2

Load the dlib face predictor.

```python
# Let's load the detector
detector = dlib.get_frontal_face_detector()

# Dlib rect --> OpenCV rect
def rect_to_bb(rect):
    
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y
    return (x, y, w, h)
```

### Step 3

Run the predictor on your image.

```python
# Init face detector by img
rects = detector(img)

# Draw rectangle around each face
for rect in rects:
    x, y, w, h = rect_to_bb(rect)
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3)
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
# Convert img color BGR to RGB for correct displaing
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Displaying detected face(s)
cv2_imshow(img)
```

### Step 5 (optional)

Repeat the process with a different and more challenging image (more faces, smaller faces, people with glasses, hats, helmets, etc.). How does the detector perform? Is it robust?

```python
# function for detecting and displaying face on img
def detect_face(img):
  rects = detector(img)

  for rect in rects:
      x, y, w, h = rect_to_bb(rect)
      cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3)
    
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  cv2_imshow(img)
```
